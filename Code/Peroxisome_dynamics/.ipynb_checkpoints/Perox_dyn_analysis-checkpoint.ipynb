#### Between day variability

```python
def plot_cell(cell):
    fig, ax = plt.subplots(1,3, figsize = (18,6))
    ax[0].plot(day1.loc[day1['Cell'] == cell]['time'], day1.loc[day1['Cell'] == cell]['count'])
    ax[1].plot(day2.loc[day2['Cell'] == cell]['time'], day2.loc[day2['Cell'] == cell]['count'])
    ax[2].plot(day3.loc[day3['Cell'] == cell]['time'], day3.loc[day3['Cell'] == cell]['count'])
    
    ylower = min(min(day1.loc[day1['Cell'] == cell]['count']),
                 min(day2.loc[day2['Cell'] == cell]['count']),
                 min(day3.loc[day3['Cell'] == cell]['count'])) - 20
    yupper = max(max(day1.loc[day1['Cell'] == cell]['count']),
             max(day2.loc[day2['Cell'] == cell]['count']),
             max(day3.loc[day3['Cell'] == cell]['count'])) + 20
    
    ax[0].set_ylim(ylower,yupper)
    ax[1].set_ylim(ylower,yupper)
    ax[2].set_ylim(ylower,yupper)

    ax[0].set_title('Day1')
    ax[1].set_title('Day2')
    ax[2].set_title('Day3')

    ax[0].set_ylabel('Count')
    ax[1].set_xlabel('Time')
    plt.suptitle('Cell{0} Count'.format(cell))
```

```python
plot_cell(1)
```

```python
plot_cell(2)
```

```python
fig, ax = plt.subplots(1,3, figsize = (18,6))
sns.boxplot(y = day1.loc[:, 'count'], ax = ax[0])
sns.boxplot(y = day2.loc[:, 'count'], ax = ax[1])
sns.boxplot(y = day3.loc[:, 'count'], ax = ax[2])

ylower = min(min(day1.loc[:, 'count']),
             min(day2.loc[:, 'count']),
             min(day3.loc[:, 'count'])) - 20
yupper = max(max(day1.loc[:, 'count']),
         max(day2.loc[:, 'count']),
         max(day3.loc[:, 'count'])) + 20

ax[0].set_ylim(ylower,yupper)
ax[1].set_ylim(ylower,yupper)
ax[2].set_ylim(ylower,yupper)

ax[0].set_title('Day1')
ax[1].set_title('Day2')
ax[2].set_title('Day3')

ax[0].set_ylabel('Count')
plt.suptitle('Count Distribution')
```

#### Between cell variability

```python
fig, ax = plt.subplots(figsize = (18,6))
all_days = pd.concat([day1, day2, day3], keys=['day1', 'day2', 'day3']).reset_index().iloc[:,[0,2,3,4]]
sns.boxplot(x = all_days['Cell'], y = all_days['count'], ax = ax)
ax.set_title('Between Cell Dist')
```

#### Normailize between day variability

```python
## Calculate median of each day
all_days = pd.concat([day1, day2, day3], keys=['day1', 'day2', 'day3']).reset_index().iloc[:,[0,2,3,4]]
day_diff = all_days.groupby('level_0')['count'].median() - all_days.loc[:, 'count'].median()
day1.loc[:, 'count'] = day1.loc[:, 'count'] - day_diff['day1']
day2.loc[:, 'count'] = day2.loc[:, 'count'] - day_diff['day2']
day3.loc[:, 'count'] = day3.loc[:, 'count'] - day_diff['day3']
```

```python
fig, ax = plt.subplots(1,3, figsize = (18,6))
sns.boxplot(y = day1.loc[:, 'count'], ax = ax[0])
sns.boxplot(y = day2.loc[:, 'count'], ax = ax[1])
sns.boxplot(y = day3.loc[:, 'count'], ax = ax[2])

ylower = min(min(day1.loc[:, 'count']),
             min(day2.loc[:, 'count']),
             min(day3.loc[:, 'count'])) - 20
yupper = max(max(day1.loc[:, 'count']),
         max(day2.loc[:, 'count']),
         max(day3.loc[:, 'count'])) + 20

ax[0].set_ylim(ylower,yupper)
ax[1].set_ylim(ylower,yupper)
ax[2].set_ylim(ylower,yupper)

ax[0].set_title('Day1')
ax[1].set_title('Day2')
ax[2].set_title('Day3')

ax[0].set_ylabel('Count')
plt.suptitle('Count Distribution')
```

#### Normailize between cell variability

```python
all_days['Cell'].unique()
```

```python
## Calculate median of each day
all_days = pd.concat([day1, day2, day3], keys=['day1', 'day2', 'day3']).reset_index().iloc[:,[0,2,3,4]]
cell_diff = all_days.groupby('Cell')['count'].median() - all_days.loc[:, 'count'].median()
con_list = list()
for cell in all_days['Cell'].unique():
    temp_df = all_days.loc[all_days['Cell'] == cell]
    temp_df.loc[:, 'count'] = temp_df.loc[:, 'count'] - cell_diff[cell]
    con_list.append(temp_df)
all_days = pd.concat(con_list)
```

```python
fig, ax = plt.subplots(figsize = (18,6))
sns.boxplot(x = all_days['Cell'], y = all_days['count'], ax = ax)
ax.set_title('Between Cell Dist')
```
